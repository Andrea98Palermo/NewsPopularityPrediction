```python
# write here the code for 2.1 Load the Dataset

import csv
import pandas as pd
from enum import Enum
import numpy as np

#First we analyze the datasetin order to clean it

class Type(Enum):
    STR = 1
    FLOAT = 2
    INT = 3


def checkTypes(row, n_row):
    """
    checks if all attributes in a row can be casted to the correct type
    """
    types = [Type.STR, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.INT,
            Type.INT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT,
            Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT,
            Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT,
            Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT,
            Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.FLOAT, Type.INT]
    
    correct = True
    for i in range(len(types)):

        try:
            if(types[i] == Type.FLOAT):         
                float(row[i])                   #check if row[i] can be casted to float
            elif(types[i] is Type.INT): 
                if float(row[i]) % 1 != 0:      #check if row[i] can be casted to float (works also for integers) and it has no decimals
                    correct = False
        except:
            correct = False
            print("Type mismatch in row #", n_row, ", field #", i)
            print("field #", i, "should be of type", types[i], "but is actually ", type(row[i]), "( value =", row[i], ")")
        
    return correct

input = open('OnlineNewsPopularity.csv', 'rt')

i=0
for row in csv.reader(input):                   #check number of attributes and types for each record
    
    if (len(row) != 61):
        print("Record #", i, " has ", len(row), "attributes")
        if(len(row) > 61):
            print("-->", row)
        print()
        print()
    
    elif (i != 0 and not checkTypes(row, i)):
        print()

    i+=1

input.close()
```

```python
#now we can use the findings we made to clean the dataset

input = open('OnlineNewsPopularity.csv', 'rt')
output = open('clean.csv', 'w')
writer = csv.writer(output)
i=0
for row in csv.reader(input):
  if i == 39536:
    del row[6]    #delete ' ' field from row 39536 which has 62 attributes
  elif i in [36262, 36263, 37910, 39591, 39592, 39644, 39645, 39647, 39648]:
    i+=1
    continue    #delete other rows (they all have less than 61 attributes excpet from row 37910 which has shares value = n.a.)
  writer.writerow(row)
  i+=1

input.close()
output.close()
```

```python
#finally we can read the cleaned csv with Pandas and substitute NA values using K-Nearest_Neighbor imputer

from sklearn.impute import KNNImputer

def log_na(df, name):
    """
    utility function to check if there are n.a. values
    """
    print("NA values present in " + name if df.isnull().values.any() else "NA values not present in " + name)

df = pd.read_csv("clean.csv", na_values = " n.a.")

X = df.drop("url", 1).drop(" shares", 1)        #create X dropping target variable and url
y = df[" shares"]       # y = target variable
X_columns = X.columns

print("Before imputing:")
log_na(X, "X")
log_na(y, "y")
print()

imputer = KNNImputer(n_neighbors=3, weights="uniform")      #impute only X as y doesn't contain n.a. values
X = pd.DataFrame(imputer.fit_transform(X))
X = X.set_axis(X_columns, axis=1, inplace=False)

print("After imputing: ")
log_na(X, "X")
log_na(y, "y")
```

```python
"""
Even if they are required in the next assignment (assignment 3: Model selection), I perform some of the steps (scaling, discretiziong of target, dataset splitting)
here because I believe the feature importance analysis could perform better using scaled features and discretized target; also, as far as I know, it is common
practice to avoid using the test set for feature importance analysis if its results will be used also for feature selection (as I will try to do) in order to avoid 
information leakage from the test set (to evaluate the models correctly their tuning has to be done without seeing any data from the test set that will be used for evaluation),
so i need to have the dataset split.
"""
```

```python

#after dropping the urls, there are only binary and numerical features in X, so there is no need of encoding any feature

bin_columns = [' data_channel_is_lifestyle', ' data_channel_is_entertainment', ' data_channel_is_bus', ' data_channel_is_socmed', ' data_channel_is_tech', ' data_channel_is_world', ' weekday_is_monday', ' weekday_is_tuesday', ' weekday_is_wednesday', ' weekday_is_thursday', ' weekday_is_friday', ' weekday_is_saturday', ' weekday_is_sunday', ' is_weekend']
num_columns = [c for c in X_columns if c not in bin_columns]
```

```python
#boxplot X to observe possible outliers

plt.rcParams["figure.figsize"] = (20,10)
plt.boxplot(X[num_columns])
plt.show()
```

```python
#As the boxplot shows there are some possible outliers, even if not explicitly required by any assignment some of them will be detected and removed

from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM

print("X's shape before removing outliers --> ", X.shape)
print("y's shape before removing outliers --> ", y.shape)

X["anomaly"] = IsolationForest(random_state=28).fit_predict(X[num_columns])
                #OneClassSVM().fit_predict(X)
                #LocalOutlierFactor(n_jobs=-1).fit_predict(X)

X_y = pd.concat([X, y], axis=1)
X_y_clean = X_y[X_y['anomaly'] != -1].drop('anomaly', axis=1)
y = X_y_clean[" shares"]
X = X_y_clean.drop(" shares", axis=1)

print("X's shape after removing outliers --> ", X.shape)
print("y's shape after removing outliers --> ", y.shape)
```

```python
#boxplot X again to observe differences between now and before deleting outliers: it seems some of them actually disappeared (e.g column 14)

plt.boxplot(X[num_columns])
plt.show()
```

```python
#feature scaling: the commented sclers have also been tested and at the end MinMaxScaler has been chosen as it resulted in better classification

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler #, StandardScaler, RobustScaler, QuantileTranformer

scaler = MinMaxScaler()     #StandardScaler(), RobustScaler(), QuantileTranformer(random_state=28, n_quantiles=300, output_distribution='normal')

preprocessor = ColumnTransformer([('standard_scaler', scaler, num_columns)], remainder='passthrough')       #scaling only numerical columns

X_scaled = pd.DataFrame(preprocessor.fit_transform(X))
X_scaled_columns = num_columns + bin_columns
X_scaled = X_scaled.set_axis(X_scaled_columns, axis=1, inplace=False)
```

```python
#boxplot X_scaled to check everything is as expected

plt.boxplot(X_scaled[num_columns])
plt.show()
```

```python
#discretization of target variable using 5 bins

from sklearn.preprocessing import KBinsDiscretizer

y = KBinsDiscretizer(n_bins=5, encode="ordinal").fit_transform(pd.DataFrame(y))
```

```python
#stratified split of dataset: full train set will be used for cross-validation

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, stratify=y, random_state=28)
```

```python
def show_importances(importances):
    """
    utility function to plot a rank of feature importances
    (sorted to easily observe elbows)
    """
    rng = range(len(importances))
    plt.xticks(rng)
    plt.grid()
    plt.plot(rng, sorted(importances, reverse=True), marker="o")
```

```python
#1st importance score: correlation of features to shares

X_train["shares"] = y_train
correlation = X_train.corr()["shares"][:-1]
X_train = X_train.drop('shares', axis=1)
show_importances(correlation)
```

```python
#2nd importance scores: random forest importances

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_jobs=-1)
clf.fit(X_train, y_train.ravel())
rf_importances = clf.feature_importances_
show_importances(rf_importances)
```

```python
#3rd importance scores: permutation importances

from sklearn.inspection import permutation_importance

result = permutation_importance(
    clf, X_test, y_test.ravel(), n_repeats=10, random_state=28, n_jobs=-1
)
perm_importances = result.importances_mean
show_importances(perm_importances)
```

```python
#4th importance scores: variances

from sklearn.feature_selection import VarianceThreshold

variances = VarianceThreshold().fit(X_train).variances_
show_importances(variances)
```

```python
#5th importance scores: ANOVA F-values

from sklearn.feature_selection import SelectPercentile

anova_importances = SelectPercentile().fit(X_train, y_train.ravel()).scores_
show_importances(anova_importances)
```

```python

# attempt of feature selection: various combinations of importances tested 
# but all of them decreased classification performance so in the end no selection was performed (more in final report)

"""
tresholds found for each importance score trying to detect elbow points in the plotted curves 
"""
corr_tresh = 0.099
rf_tresh = sorted(rf_importances, reverse=True)[42]
perm_tresh = sorted(perm_importances, reverse=True)[23]
var_tresh = sorted(variances, reverse=True)[31]
anova_tresh = sorted(anova_importances, reverse=True)[44]


print("X_train's shape before feature selection --> ", X_train.shape)
print("X_test's shape before feature selection --> ", X_test.shape)


"""
Only features with importance above treshold for all (or a subset of) the importance scores are kept.
Best classification performances were obtained keeping all features
"""
new_features = [col for i, col in enumerate(X_scaled_columns) #if 
                #rf_importances[i] >= rf_tresh
                #and np.abs(correlation[i]) >= corr_tresh
                #and perm_importances[i] >= perm_tresh
                #and variances[i] >= var_tresh
                #anova_importances[i] >= anova_tresh
              ]
X_train_reduced = X_train[new_features]
X_test_reduced = X_test[new_features]

print("X_train's shape after feature selection --> ", X_train_reduced.shape)
print("X_test's shape after feature selection --> ", X_test_reduced.shape)
```

your comments for step 2.4 Feature importance analysis

I used five different methods, as seen above, in order to get an importance score for each feature: correlation to shares, random forest classifier's importance, permutation importance, variance and ANOVA F-value (a score given by inter_group_variance / intra_group_variance).
After ranking the features by each type of importance, I plotted the sorted importances in order to try to find elbow points in the curves (to use them as feature selection tresholds). Once I defined these tresholds I made some attempts to do feature selection using various combination of the tresholds: I kept only the features with each (selected) type of importance greather than its relative treshold.
Unfortunately, none of the attempts of using different combinations of the importances managed to improve the predictions quality (next step), so in the end no feature selection was performed.

```python
#definition of functions to tune hyper-parameters and evaluate models

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
from sklearn.metrics import classification_report, f1_score, accuracy_score

def search_report(clf, X_test, y_test, score):
    """
    prints a report of the performed grid/random search
    and evaluates best model on unseen test set
    """
    print("Best parameters found using 3-fold cross validation on X_train:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on X_train:")
    print()

    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
            % (mean, std * 2, params))
    print()

    y_true, y_pred = y_test, clf.predict(X_test)
    print("Classification report of model using best parameters tested on unseen X_test:")
    print()
    print(classification_report(y_true, y_pred))
    print()

    print("F1_macro score of model using best parameters tested on unseen X_test:")
    print(f1_score(y_true, y_pred, average="macro"))
    print()

    print("Accuracy score of model using best parameters tested on unseen X_test:")
    print(accuracy_score(y_true, y_pred))
    print()



def grid_search_tuning(tuned_parameters, scores, model, X_train, y_train, X_test, y_test):
    """
    performs a grid search using given model, parameters grid and dataset
    """
    for score in scores:
        print("# Tuning hyper-parameters for ",  score)
        print()

        clf = GridSearchCV(
            model, tuned_parameters, scoring=score, cv=3
        )
        clf.fit(X_train, y_train)

        search_report(clf, X_test, y_test, score)


def random_search_tuning(tuned_parameters, scores, model, iterations, X_train, y_train, X_test, y_test):
    """
    performs a random search using given model, parameters grid and dataset
    """
    for score in scores:
        print("# Tuning hyper-parameters for", score)
        print()

        clf = RandomizedSearchCV(
            model, tuned_parameters, scoring=score, cv=3, n_iter=iterations
        )
        clf.fit(X_train, y_train)

        search_report(clf, X_test, y_test, score)
```

The principal finding I made while carrying out the required dataset analysis tasks and plots is the skewed distribution of the target feature, and I realized this could lead to having unbalanced classes when discretizing the target variable.
